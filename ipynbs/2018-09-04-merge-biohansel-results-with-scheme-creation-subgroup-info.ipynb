```python
df_scheme_creation = pd.read_csv('SH1002-subtype-groups-min-group-size-5-2snv-all-methods.csv', index_col=0)
```

```python
df_scheme_creation
```

```python
df_scheme_creation['sample'] = df_scheme_creation.index
```

```python
df_bh_sh_subtype = df_bh_sh[['sample', 'subtype']]
```

```python
pd.DataFrame(list(df_bh_sh_subtype.subtype.str.split(r'\.')))
```

```python
subtype_at_levels = []
for r in df_bh_sh_subtype.subtype.str.split(r'\.'):
    st = ''
    tmp = []
    for x in r:
        if st == '':
            st += x
        else:
            st += '.' + x
        tmp.append(st)
    subtype_at_levels.append(tmp)
```

```python
df_bh_sh_at_diff_levels = pd.DataFrame(subtype_at_levels)
```

```python
df_bh_sh_at_diff_levels.index = df_bh_sh_subtype['sample']
```

```python
df_bh_sh_at_diff_levels.columns = [f'L{c + 1}:o' for c in df_bh_sh_at_diff_levels.columns]
    
```

```python
df_scheme_creation = df_scheme_creation.drop(columns=['sample'])
```

```python
df_merge = pd.concat([df_bh_sh_at_diff_levels, df_scheme_creation], axis=1)
```

```python
df_merge
```

```python
df_merge.to_csv('SH1002-biohansel-subtype-scheme-creation-subtype-groups-min-group-size-5-2snv-all-methods.csv')
```
